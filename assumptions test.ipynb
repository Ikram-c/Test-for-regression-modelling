```python
df = df.drop(['Notes','Status','Agency','Benefits','Id'], axis=1)
```

```python
df = df.drop(df[df.EmployeeName == 'Not provided'].index)
df = df.drop(df[df.JobTitle == 'Not provided'].index)
df = df.drop(df[df.BasePay == 'Not provided'].index)
df = df.drop(df[df.OvertimePay == 'Not provided'].index)
df = df.drop(df[df.OtherPay == 'Not provided'].index)
df = df.drop(df[df.TotalPay == 'Not provided'].index)
df = df.drop(df[df.TotalPayBenefits == 'Not provided'].index)
df = df.drop(df[df.Year == 'Not provided'].index)
```

```python
import scipy
from scipy.stats.stats import pearsonr

def pairplot(x_axis,y_axis) :
    sns.pairplot(df,x_vars=x_axis,y_vars=y_axis,height=4,aspect=1,kind="scatter")
    plt.show()
```

```python
df = df.dropna()
```

```python
def heatmap(x,y,dataframe):
    sns.heatmap(dataframe.corr(),cmap="OrRd",annot=True)
    plt.show()
heatmap(20,12,df)
```

```python
y_vars=['TotalPay']
```

```python
x_vars=['BasePay','OvertimePay','TotalPayBenefits']
pairplot(x_vars,y_vars)

p1=df['BasePay']
p2=df['OvertimePay']
p3=df['TotalPayBenefits']
pearson_coeff, p_value = pearsonr(p1,df['TotalPay'])
print('\nWe can make sure of this by looking at the Coefficient of Correlation')
print('\nCoefficient of Correlation between TotalPay and BasePay:',pearson_coeff*100,'%')

pearson_coeff, p_value = pearsonr(p2,df['TotalPay'])
print('Correlation coefficient between TotalPay and OvertimePay:',pearson_coeff*100,'%')

pearson_coeff, p_value = pearsonr(p3,df['TotalPay'])
print('Correlation coefficient between Price and TotalPayBenefits: ',pearson_coeff*100,'%')
```
